```python
import torch
from torch.utils.data import Dataset
from PIL import Image

class IAMDataset(Dataset):
    def __init__(self, root_dir, df, feature_extractor,tokenizer, max_target_length=128):
        self.root_dir = root_dir
        self.df = df
        self.feature_extractor = feature_extractor
        self.tokenizer = tokenizer
        self.max_target_length = max_target_length
    def __len__(self):
        return len(self.df)
    def __getitem__(self, idx):
        file_name = self.df['file_name'][idx]
        text = self.df['text'][idx]
        image = Image.open(self.root_dir + file_name).convert("RGB")
        pixel_values = self.feature_extractor(image, return_tensors="pt").pixel_values
        labels = self.tokenizer(text, padding="max_length", max_length=self.max_target_length,truncation =True).input_ids
        labels = [label if label != self.tokenizer.pad_token_id else -100 for label in labels]
        encoding = {"pixel_values": pixel_values.squeeze(), "labels": torch.tensor(labels)}
        return encoding
```

```python
!pip install python-Levenshtein jiwer
!pip install evaluate
from evaluate import load

# Load the CER metric
cer_metric = load("cer")
from jiwer import wer

def compute_wer(pred_str, label_str):
    errors = [wer(label, pred) for pred, label in zip(pred_str, label_str)]
    avg_wer = sum(errors) / len(errors)
    return avg_wer

from Levenshtein import distance as levenshtein_distance

def compute_levenshtein_distance(pred_str, label_str):
    distances = [levenshtein_distance(pred, label) for pred, label in zip(pred_str, label_str)]
    avg_distance = sum(distances) / len(distances)
    return avg_distance
```

```python
def compute_metrics(pred):
    labels_ids = pred.label_ids
    pred_ids = pred.predictions

    pred_str = decoder_tokenizer.batch_decode(pred_ids, skip_special_tokens=True)
    labels_ids[labels_ids == -100] = decoder_tokenizer.pad_token_id
    label_str = decoder_tokenizer.batch_decode(labels_ids, skip_special_tokens=True)

    cer = cer_metric.compute(predictions=pred_str, references=label_str)
    levenshtein = compute_levenshtein_distance(pred_str, label_str)

    # Compute WER
    wer_score = compute_wer(pred_str, label_str)

    return {
        "cer": cer,
        "levenshtein_distance": levenshtein,
        "wer": wer_score,
    }
```

```python
import os
import random
from PIL import Image
import torch
import matplotlib.pyplot as plt
from jiwer import cer  # Character Error Rate (CER) computation
from matplotlib import rcParams
from matplotlib import font_manager

# Path to the downloaded font file
font_path = '/content/drive/MyDrive/Colab Notebooks/Character Recognition organized/NotoSansBengali-VariableFont_wdth,wght.ttf'

font_dirs = [os.path.dirname(font_path)]
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

# Set the font family globally
rcParams['font.family'] = 'Noto Sans Bengali'



# Iterate through the specified range of images
for i in range(50):
    file_name = f"{random.randint(1,1999)}.jpg"
    image_path = os.path.join('/content/drive/MyDrive/Colab Notebooks/Character Recognition organized/train/', file_name)

    if not os.path.exists(image_path):
        print(f"File {file_name} not found. Skipping...")
        continue

    # Open and preprocess the image
    image = Image.open(image_path).convert("RGB")
    pixel_values = processor.feature_extractor(image, return_tensors="pt").pixel_values

    # Move to GPU if available
    if torch.cuda.is_available():
        pixel_values = pixel_values.cuda()
        model = model.cuda()

    # Generate prediction
    generated_ids = model.generate(pixel_values)
    predicted_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    # predicted_text = correct_predictions(predicted_text)

    # Get the ground truth from the dataframe
    ground_truth = df[df['file_name'] == file_name]['text'].values
    if len(ground_truth) == 0:
        print(f"Ground truth not found for {file_name}. Skipping...")
        continue

    ground_truth = ground_truth[0]

    # Calculate Character Error Rate (CER)
    cer_value = cer(ground_truth, predicted_text)

    # Calculate Accuracy (percentage of correctly matched characters)
    correct_chars = sum(1 for gt, pred in zip(ground_truth, predicted_text) if gt == pred)
    total_chars = len(ground_truth)
    accuracy = (correct_chars / total_chars) * 100 if total_chars > 0 else 0

    # Display the image and print results
    plt.imshow(image)
    plt.axis('off')
    plt.title(f"File: {file_name}\nPredicted: {predicted_text}\nActual: {ground_truth}\nCER: {cer_value:.4f} | Accuracy: {accuracy:.2f}%")
    plt.show()


```
